```python
import pandas as pd
import numpy as np
df = pd.read_csv("video_game_sales.csv")
```

```python
#Dataset overview
print(df.shape)
print(df.dtypes)
print(df.describe(include="all"))
print(df.isnull().sum())
```

```python
#Drop empty year entries and reassign dataypes
df = df.dropna(subset=['Year'])
df['Year'] = df['Year'].astype('int64')
object_col = df.select_dtypes(include=['object']).columns
df[object_col] = df[object_col].astype('string')

#Round float data types to 2 decimal places
float_col = df.select_dtypes(include=['float64']).columns
df[float_col] = df[float_col].round(2)
```

```python
#Save and check new data
df.to_csv("new_video_game_sales.csv", index=False)
print(df.shape)
print(df.dtypes)
print(df.describe(include="all"))
print(df.isnull().sum())
```
